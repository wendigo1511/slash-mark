```python
import colorama
from colorama import Fore, Back, Style

# Define an empty list to store tasks
tasks = []

# Function to display the to-do list
def display_tasks():
    if not tasks:
        print(Fore.RED + "Your to-do list is empty." + Style.RESET_ALL)
    else:
        print(Fore.BLUE + "To-Do List:" + Style.RESET_ALL)
        for i, task in enumerate(tasks, start=1):
            status = "Done" if task["completed"] else "Not Done"
            print(f"{Fore.GREEN}{i}. {task['task']} ({status}){Style.RESET_ALL}")

# Function to add a task to the to-do list
def add_task(task_name):
    task = {"task": task_name, "completed": False}
    tasks.append(task)
    print(Fore.GREEN + f"Task '{task_name}' added to your to-do list." + Style.RESET_ALL)

# Function to mark a task as completed
def mark_completed(task_number):
    if 1 <= task_number <= len(tasks):
        tasks[task_number - 1]["completed"] = True
        print(Fore.GREEN + f"Task {task_number} marked as completed." + Style.RESET_ALL)
    else:
        print(Fore.RED + "Invalid task number. Please enter a valid task number." + Style.RESET_ALL)

# Function to remove a task from the to-do list
def remove_task(task_number):
    if 1 <= task_number <= len(tasks):
        removed_task = tasks.pop(task_number - 1)
        print(Fore.GREEN + f"Task '{removed_task['task']}' removed from your to-do list." + Style.RESET_ALL)
    else:
        print(Fore.RED + "Invalid task number. Please enter a valid task number." + Style.RESET_ALL)

# Main program loop
while True:
    print(Fore.CYAN + "\nOptions:" + Style.RESET_ALL)
    print("1. Display to-do list")
    print("2. Add a task")
    print("3. Mark a task as completed")
    print("4. Remove a task")
    print("5. Quit")
    choice = input("Enter your choice: ")

    if choice == '1':
        display_tasks()
    elif choice == '2':
        task_name = input("Enter the task: ")
        add_task(task_name)
    elif choice == '3':
        display_tasks()
        task_number = int(input("Enter the task number to mark as completed: "))
        mark_completed(task_number)
    elif choice == '4':
        display_tasks()
        task_number = int(input("Enter the task number to remove: "))
        remove_task(task_number)
    elif choice == '5':
        break
    else:
        print(Fore.RED + "Invalid choice. Please enter a valid option." + Style.RESET_ALL)
```
